```python
# 决定threshold
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
import numpy as np
def plot_precision_recall_vs_threshold(y_train, y_scores):
    precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)
    print(precisions, recalls, thresholds)
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.xlabel("Threshold")
    plt.legend(loc="upper left")
    plt.ylim([0, 1])
    plt.show()

y_scores = np.array([1,0,1,0,1,0,0,0,1,1])
np.random.seed(1234)
y_train = np.random.random(10)

plot_precision_recall_vs_threshold(y_scores, y_train)
```

```python
# ROC曲线
from sklearn.metrics import roc_curve,roc_auc_score

def plot_roc_curve(y_train, y_scores, label=None):
    fpr, tpr, thresholds = roc_curve(y_train, y_scores)
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
plot_roc_curve(y_scores, y_train)
plt.show()
auc = roc_auc_score(y_scores, y_train)
print("auc:{}".format(auc))
```
